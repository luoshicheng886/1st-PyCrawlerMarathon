```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response)
table = soup.find('table', class_='TABLE_G')
# print(table)
```

```python
aq_dict = {}
main_key = ''
tr_list = table.find_all('tr')
for tr in tr_list:
    td_list = tr.find_all('td')
    if len(td_list) > 3 and td_list[4].text != '無此測項':
        main_key = td_list[0].text
        aq_dict[main_key] = {td_list[2].text: td_list[3].text}
    elif len(td_list) == 3:
        aq_dict[main_key][td_list[0].text] = td_list[1].text
print(aq_dict)
```
