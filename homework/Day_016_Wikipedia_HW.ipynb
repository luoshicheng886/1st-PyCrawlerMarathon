# Wikipedia爬蟲練習
## 範例：練習是從Wikipedia中爬取文章。先定義一個搜尋的關鍵字，擷取該關鍵字詞的文章。

```python
import urllib
import gzip
import re
from bs4 import BeautifulSoup
```

### 先定義一個我們想搜尋的字詞，並將它轉換成UTF-8編碼後的URL

```python
input_keyword = "網路爬蟲"  # 這裡可以自己定義有興趣的關鍵字
input_keyword = input_keyword.replace(' ', '_')
utf8_url = urllib.parse.quote(input_keyword)
print("%s: %s" % (input_keyword, utf8_url))
# 組成Wiki關鍵字搜尋的網址格式
root_keyword_link = '/wiki/' + utf8_url
print(root_keyword_link)
```

### 範例1：送出關鍵字請求後，爬取該關鍵字的文章內容

```python
# 模擬封包的標頭
headers = {
    'authority': 'zh.wikipedia.org',
    'method': 'GET',
    'path': '/wiki/' + root_keyword_link,
    'scheme': 'https',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6',
    'cookie': 'GeoIP=TW:TPE:Taipei:25.05:121.53:v4; TBLkisOn=0; mwPhp7Seed=8b8; WMF-Last-Access-Global=04-Jun-2019; WMF-Last-Access=04-Jun-2019',
    'dnt': '1',
    #'if-modified-since': 'Tue, 04 Jun 2019 12:03:22 GMT',
    'referer': 'https://zh.wikipedia.org/wiki/Wikipedia:%E9%A6%96%E9%A1%B5',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}    

url = 'https://zh.wikipedia.org' + root_keyword_link  # 組合關鍵字查詢URL
request = urllib.request.Request(url, headers=headers)
response = urllib.request.urlopen(request)
response = gzip.decompress(response.read())
html = BeautifulSoup(response.decode('utf-8'))
content = html.find(name='div', attrs={'id':'mw-content-text'}).find_all(name='p')
#
# 解析回傳資料，並萃取文章內容
#
for p_tag in content:
    print(p_tag.text)
```

### 範例2：從爬取的文章內容中，擷取出有外部連結的關鍵字。這些關鍵字在文章中是以藍色字體顯示，會連到外部的網頁，並解釋其內容。

```python
for p_tag in content:
    a_tags = p_tag.find_all('a', href=re.compile("^(/wiki/)((?!;)\S)*$"))
    if len(a_tags) > 0:
        for a in a_tags:
            a_href = a["href"]  # 外部連結的網址
            a_keyword = a.text  # 外部連結的中文名稱
            print("外部連結: [%s] %s" % (a_keyword, urllib.parse.unquote(a_href)))
```

## 作業：接下來定義一個爬蟲函數，這個函數的主要工作為：
### (1) 爬取當前關鍵字的解釋，並存入檔案(因為文章內容太多會佔滿整個頁面，所以存程檔案，方便後續檢視)
### (2) 萃取出當前關鍵字所引用的外部連結，當作新的查詢關鍵字
### (3) 把第(2)擷取到的關鍵字當作新的關鍵字，回到第(1)步，爬取新的關鍵字解釋。

```python
import os

# 定義爬取的遞迴深度。深度不要訂太深，否則會爬很久。
max_recursive_depth = 2

def WikiArticle(keyword_link, keyword, recursive):

    if recursive <= max_recursive_depth:
        print("遞迴層[%d] - %s (%s)" % (recursive, urllib.parse.unquote(keyword_link), keyword))

        # 模擬封包的標頭
        headers = {
            'authority': 'zh.wikipedia.org',
            'method': 'GET',
            'path': '/wiki/' + keyword_link,
            'scheme': 'https',
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
            'accept-encoding': 'gzip, deflate, br',
            'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6',
            'cookie': 'GeoIP=TW:TPE:Taipei:25.05:121.53:v4; TBLkisOn=0; mwPhp7Seed=8b8; WMF-Last-Access-Global=04-Jun-2019; WMF-Last-Access=04-Jun-2019',
            'dnt': '1',
            #'if-modified-since': 'Tue, 04 Jun 2019 12:03:22 GMT',
            'referer': 'https://zh.wikipedia.org/wiki/Wikipedia:%E9%A6%96%E9%A1%B5',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
        }    

        url = 'https://zh.wikipedia.org' + keyword_link  # 組合關鍵字查詢URL
        request = urllib.request.Request(url, headers=headers)
        response = urllib.request.urlopen(request)
        response = gzip.decompress(response.read())
        html = BeautifulSoup(response.decode('utf-8'))
        content = html.find(name='div', attrs={'id':'mw-content-text'}).find_all(name='p')

        #
        # Part 1: 請參考範例1，爬取當前關鍵字的文章內容。
        #         因為內容太多，我們把它寫入檔案，並以關鍵字作為檔案名稱，以便稍後查閱內容。
        #         請先建立一個名為"WikiArticle"的資料夾，爬取到的文章內容會放在這個資料夾底下。
        #
        if not os.path.isdir('WikiArticle'):
            os.mkdir('WikiArticle', mode=0o755)
        ntfs_chrs = [chr(i) for i in range(0, 31)]
        chrs = ['\"', '\#', '\$', '\%', '\'', '\*', '\,', '\.', '\/', '\:', '"',
                '\;', '\<', '\>', '\?', '\\', '\^', '\|', '\~', '\\\\',]
        pattern = '|'.join(ntfs_chrs + chrs)
        regex = re.compile(pattern, re.UNICODE)
        with open('WikiArticle/%s' % regex.sub('', keyword), 'w') as fp:
            for p_tag in content:
                fp.write(p_tag.text)

        #
        # Part 2: 請參考範例2，萃取出本篇文章中所延伸引用的外部連結，並儲存在external_link_dict
        #
        href_dict = {}
        for p_tag in content:
            a_tags = p_tag.find_all('a', href=re.compile("^(/wiki/)((?!;)\S)*$"))
            if len(a_tags) > 0:
                for a in a_tags:
                    a_href = a["href"] # 外部連結的網址
                    a_keyword = a.text # 外部連結的中文名稱
                    href_dict[a_keyword] = a_href
                    print("\t[%s] %s" % (a_keyword, urllib.parse.unquote(a_href)))
        #
        # Part 3: 將Part 2所收集的外部連結，當作新的關鍵字，繼續迭代深入爬蟲
        #
        if len(href_dict) > 0:
            recursive += 1  # 遞迴深度加1
            for key, link in href_dict.items():
                WikiArticle(link, key, recursive)  # 再次呼叫同樣的函數，執行同樣的流程
```

### 執行前個步驟定義好的爬蟲主程式

```python
WikiArticle(root_keyword_link, input_keyword, 0)
```
